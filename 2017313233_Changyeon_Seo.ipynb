### Happy Number
Write an algorithm to determine if a number **n** is happy.

A **happy number** is a number defined by the following process:

* Starting with any positive integer, replace the number by the sum of the squares of its digits.
* Repeat the process until the number equals 1 (where it will stay), or it **loops endlessly in a cycle** which does not include 1.
* Those numbers for which this process **ends in 1** are happy.

Return **true** if n is a happy number, and **false** if not.

Constraints:
1 <= n <= $2^{31}$ - 1

```python
def get_addition_of_square_of_each_digits(num):
  result = 0
  while num != 0:
    result += ((num % 10) ** 2)
    num = num // 10
  return result

def check_cycle(num):
  result = True
  number_lists = [ num ]

  while True:
    addition_of_sqaure_of_digits = get_addition_of_square_of_each_digits(number_lists[-1])

    if addition_of_sqaure_of_digits in number_lists:
      result = True
      break
    elif addition_of_sqaure_of_digits == 1:
      result = False
      break
    else:
      number_lists.append(addition_of_sqaure_of_digits)
      continue

  return result

def isHappy(n):
  if n == 1:
    return True
  else:
    return check_cycle(n) == False
```
